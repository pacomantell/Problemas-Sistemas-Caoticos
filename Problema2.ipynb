```python
#nulclinas
def nulclinaA(a, b, c, figx=None):
    V = np.linspace(-2, 4, 30)
    omega1 = V * (a - V) * (V - 1)
    omega2 = b * V / c
    
    Vmas = 0.5 * ((a + 1) + np.sqrt((a - 1) ** 2 - 4 * (b / c)))
    Vmen = 0.5 * ((a + 1) - np.sqrt((a - 1) ** 2 - 4 * (b / c)))
    omegamas = b * Vmas / c
    omegamen = b * Vmen / c

    
    figx.plot(V, omega1, label=(r'$\dot V =0$'))
    figx.plot(V, omega2, label=(r'$\dot \omega =0$'))
    
    figx.axhline(0, color='k', lw=0.5)
    figx.set_title(f"Nulclinas para $a={a}$, $b={b}$ y $c={c}$")
    figx.plot(0, 0, 'ok', label=(f"$V_0^*=\omega_0^*=0$"))
    #figx.plot(Vmas, omegamas, 'ob', label=(f"($V_+^*$,$\omega_+^*$)"))
    #figx.plot(Vmen, omegamen, 'or', label=(f"($V_1^*$,$\omega_1^*$)"))

    
fig, fig1 =plt.subplots(1, 1, figsize=(8,4))

nulclinaA(2, 1, 1, figx=fig1)
plt.legend()
#plt.savefig('nulclina_2a3.png')
```

```python
#punto critico x0=(0,0)
def plot1(a, c, figx=None):
    #a = np.linspace(-4, 4, 50)    
    b = np.linspace(-5, 5, 50)
    #c = np.linspace(-4, 4, 50)
    
    tr = - a - c
    det = a * c + b
    delta = tr ** 2 - 4 * det
    
    lM = 0.5 * (tr + np.sqrt(delta))
    lm = 0.5 * (tr - np.sqrt(delta))
    
    bc = 0.25 * (a - c) ** 2

    
    figx.plot(b, lM, 'k', label=('$\lambda_+$'))
    figx.plot(b, lm, 'b', label='$\lambda_-$')
    figx.plot(b, delta, label=('$\Delta_0$'))
        
    figx.axhline(0, color='k', lw=0.5)
    figx.axvline(bc, color='k', lw=0.5, ls='--')
    figx.set_title(f"$a={a}, c={c}$")
    #figx.text(-3.5, -10, 'Nodo Inestable', fontsize='15', color='k')
    #figx.text(1, 5, 'Espiral Estable', fontsize='15', color='k')
    figx.set_xlabel('b')
        
fig, fig2=plt.subplots(1,1, figsize=(8,4))
plot1(-2, -1,figx=fig2)

plt.legend()
#plt.savefig('Estabilidad_2a1.png')
```

```python
#APARTADO A CAMPOS VECTORIALES
def campoA(a, b, c, figx=None):
    V1 = np.linspace(-4, 4, 30)
    omega1 = V1 * (a - V1) * (V1 - 1)
    omega2 = b * V1 / c
    
    Vmas = 0.5 * ((a + 1) + np.sqrt((a - 1) ** 2 - 4 * (b / c)))
    Vmen = 0.5 * ((a + 1) - np.sqrt((a - 1) ** 2 - 4 * (b / c)))
    omegamas = b * Vmas / c
    omegamen = b * Vmen / c

    
    figx.plot(V1, omega1, label=(r'$\dot V =0$'))
    figx.plot(V1, omega2, label=(r'$\dot \omega =0$'))
    
    figx.axhline(0, color='k', lw=0.5)
    figx.set_title(f"$a={a}$, $b={b}$ y $c={c}$")
    figx.plot(0, 0, 'ok', label=(f"$V_0^*=\omega_0^*=0$"))
    figx.plot(Vmas, omegamas, 'ob', label=(f"($V_+^*$,$\omega_+^*$)"))
    figx.plot(Vmen, omegamen, 'or', label=(f"($V_1^*$,$\omega_1^*$)"))
    
    V, O = np.meshgrid(np.linspace(-4, 4, 100), np.linspace(-4, 4, 100))
    
    Vp = (V * (a - V) * (V - 1)) - O
    Op = (b * V) - (c * O)
    
    figx.streamplot(V, O, Vp, Op, density=2)
    
    figx.set_xlim(-2,4)
    figx.set_ylim(-4,4)
    
fig, fig6 =plt.subplots(1, 1, figsize=(8,4))

campoA(2, -1, 1, figx=fig6)
#plt.legend()
#plt.savefig('espiral_estable_2a.png')
```

```python
#APARTADOB CAMPOS VECTORIALES
def campoB(a, b, c, figx=None):
    V1 = np.linspace(-0.2, 1.2, 100)
    omega1 = V1 * (a - V1) * (V1 - 1)
    omega2 = b * V1 / c
    
    Vmas = 0.5 * ((a + 1) + np.sqrt((a - 1) ** 2 - 4 * (b / c)))
    Vmen = 0.5 * ((a + 1) - np.sqrt((a - 1) ** 2 - 4 * (b / c)))
    omegamas = b * Vmas / c
    omegamen = b * Vmen / c

    
    figx.plot(V1, omega1, label=(r'$\dot V =0$'))
    figx.plot(V1, omega2, label=(r'$\dot \omega =0$'))
    
    figx.axhline(0, color='k', lw=0.5)
    figx.set_title(f"$a={a}$, $b={b:.1e}$ y $c={c}$")
    figx.plot(0, 0, 'ok', label=(f"$V_0^*=\omega_0^*=0$"))
    figx.plot(Vmas, omegamas, 'ob', label=(f"($V_+^*$,$\omega_+^*$)"))
    figx.plot(Vmen, omegamen, 'xr', label=(f"($V_1^*$,$\omega_1^*$)"))
    
    V, O = np.meshgrid(np.linspace(-0.2, 1.2, 100), np.linspace(-0.2, 1.2, 100))
    
    Vp = (V * (a - V) * (V - 1)) - O
    Op = (b * V) - (c * O)
    
    figx.streamplot(V, O, Vp, Op, density=3)
    
    figx.set_xlim(-0.2,1.2)
    figx.set_ylim(-0.2,0.2)
    
fig, fig6 =plt.subplots(1, 1, figsize=(8,4))

campoB(0.1, 0.008, 0.033, figx=fig6)
plt.savefig('sin_i_ext.png')
```

# APARTADO C

```python
#APARTADO C CAMPOS VECTORIALES
def campoC(a, b, c, I, n, figx=None):
    
    #Nulclinas
    V1 = np.linspace(-1.5, 2.2, 100)
    omega1 = (V1 * (a - V1) * (V1 - 1)) + I
    omega2 = b * V1 / c
    
    figx.axhline(0, color='k', lw=0.5)
    figx.set_title(f"$I_e ={I}$")

    figx.plot(V1, omega1, 'b', label=(r'$\dot V =0$'))
    figx.plot(V1, omega2, 'g', label=(r'$\dot \omega =0$'))
    
    #Campo vectorial
    
    V, O = np.meshgrid(np.linspace(-0.5, 1, 100), np.linspace(-0.5, 0.5, 100))
    
    Vp = (V * (a - V) * (V - 1)) - O + I
    Op = (b * V) - (c * O)
    
    figx.streamplot(V, O, Vp, Op, density=1, color='y')
    
    


    #trayectorias
    def dinamica(x0, t):
        V2 = x0[0]
        O2 = x0[1]
    
        Vp2 = (V2 * (a - V2) * (V2 - 1)) - O2 + I
        Op2 = (b * V2) - (c * O2)
    
        xp = [Vp2, Op2]
        
        return xp
    
    x0 = [0, 0]
    #hago varias trayectorias para varias condiciones iniciales
    for j in range(n):
        paso = 1 / n

        #Comdicion inicial:
        #tiempo

        t = np.linspace(0, 500, 10000)

        #almacenamiento de la solucion

        X = np.empty_like(t)
        Y = np.empty_like(t)

        X[0] = x0[0]
        Y[0] = x0[1]

#resuelvo
        for i in range(1, len(t)):
            ts = [t[i-1], t[i]]
    
            z = odeint(dinamica, x0, ts)
    
            #guardo solucion
            X[i] = z[1][0]
            Y[i] = z[1][1]
    
            x0 = z[1]
        
        figx.plot(X,Y, 'r', label=('Ciclo Limite'))
        x0 = x0 + paso
    

    figx.set_xlim(-0.5, 1)
    figx.set_ylim(-0.5, 0.5)
    figx.legend(loc=0)

fig, fig6 =plt.subplots(1, 1, figsize=(8,4))

campoC(0.1, 0.008, 0.033, 2e-2, 1, figx=fig6)

plt.savefig('trayectorias_espacio_fase.png')
```

```python
#TRAYEXTORIAS APARTADO C
def trayectorias(I, figx=None):
    def dinamica(x0, t):
        a, b, c = 0.1, 0.008, 0.033
        V = x0[0]
        O = x0[1]
    
        Vp = (V * (a - V) * (V - 1)) - O + I
        Op = (b * V) - (c * O)
    
        xp = [Vp, Op]
        
        return xp
    
    #Comdicion inicial:
    x0 = [0, 0]
    
    #tiempo

    t = np.linspace(0, 1000, 10000)

    #almacenamiento de la solucion

    X = np.empty_like(t)
    Y = np.empty_like(t)

    X[0] = x0[0]
    Y[0] = x0[1]

    #resuelvo
    for i in range(1, len(t)):
        ts = [t[i-1], t[i]]

        z = odeint(dinamica, x0, ts)
        
        #guardo solucion
        X[i] = z[1][0]
        Y[i] = z[1][1]
    
        x0 = z[1]
    
    figx.plot(t, X, label=('V(t)'))
    figx.plot(t, Y, label=('$\omega(t)$'))
    
    figx.set_xlabel('t')
    figx.set_title(f'$I={I:0.1e}$')
    figx.legend(loc=0)
fig, fig8 = plt.subplots(1,1,figsize=(8,4))

trayectorias(3.2e-2, figx=fig8)

plt.savefig('trayectorias.png')
```

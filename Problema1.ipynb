```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# APARTADO A

```python
#APARTADO A
def amor1(a, b, R, L):
    return [a * L, b * R]
```

```python
#nulclinas
def nulclinaA(figx=None):
    figx.axhline(0, color='r', lw='2', label=r'$\dot L =0$')
    figx.axvline(0, color='b', lw='2', label=r'$\dot R =0$')
    figx.plot(0,0, 'ok', label=('Punto Fijo'))
    figx.set_title('Nulclinas apartado a')
    figx.set_xlabel('R')
    figx.set_ylabel('L')
    figx.set_xlim(-2,2)
    figx.set_ylim(-2,2)
fig, fig1=plt.subplots(1,1, figsize=(8,4))
nulclinaA(figx=fig1)
plt.legend()
plt.savefig('nulclina_1a.png')
```

```python
#campo vectorial a
def campoA(a, b, figx=None):
    figx.axhline(0, color='r', lw='2', label=r'$\dot L =0$')
    figx.axvline(0, color='b', lw='2', label=r'$\dot R =0$')
    figx.plot(0,0, 'ok', label=('Punto Fijo'))
    figx.set_title(f'$a={a}, b={b}$')
    figx.set_xlabel('R')
    figx.set_ylabel('L')
    figx.set_xlim(-2,2)
    figx.set_ylim(-2,2)
    
    
    R, L = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))
    
    Rpunto = a * L
    Lpunto = b * R

    
    
    #figx.quiver(R, L, Rpunto, Lpunto)
    figx.streamplot(R, L, Rpunto, Lpunto, density=1)

    
fig, fig5=plt.subplots(1,1, figsize=(8,4))
campoA(1, 1, figx=fig5)

plt.savefig('saddle_point_1a.png')
```

```python
#trayectorias

def plot2(a, b, figx=None):
    t = np.linspace(0, 10, 50)
    L = (1 / (a * b)) * np.exp(a * b * t)
    R = (1 / (a * b ** 2)) * np.exp(a * b * t)
    
    figx.plot(t, L, 'k')
    figx.plot(t, R, 'b')
    
fig, fig2=plt.subplots(1, 1, figsize=(12,6))

plot2(1, -1, figx=fig2)
```

# APARTADO B

```python
#nulclinas apartado b
def nulclinaB(a, b, figx=None):
    R = np.linspace(-2, 2, 20)
    k = a / b
    L1 = - k * R
    L2 = - R / k
    figx.plot(R, L1, 'r', lw='2', label=(r'$\dot R =0$'))
    figx.plot(R, L2, 'b', lw='2', label=(r'$\dot L =0$'))
    figx.plot(0, 0, 'ok', label=('Punto Fijo'))
    figx.set_title(f"Nulclinas para $a={a}$ y $b={b}$")
    
fig, fig3=plt.subplots(1, 1, figsize=(8,4))
nulclinaB(2, 1, figx=fig3)
plt.legend()
plt.savefig('nulclina_1b.png')
    
```

```python
#campo vectorial b
def campoB(a, b, figx=None):
    R1 = np.linspace(-2, 2, 20)
    k = a / b
    L1 = - k * R1
    L2 = - R1 / k
    figx.plot(R1, L1, 'r', lw='2', label=(r'$\dot R =0$'))
    figx.plot(R1, L2, 'b', lw='2', label=(r'$\dot L =0$'))
    figx.plot(0, 0, 'ok', label=('Punto Fijo'))
    
    figx.set_title(f'$a={a}, b={b}$')
    figx.set_xlabel('R')
    figx.set_ylabel('L')
    figx.set_xlim(-2,2)
    figx.set_ylim(-2,2)
    
    
    R, L = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))
    
    Rpunto = (a * R) + (b * L)
    Lpunto = (b * R) + (a * L)

    
    
    #figx.quiver(R, L, Rpunto, Lpunto)
    figx.streamplot(R, L, Rpunto, Lpunto, density=1)
    
    if (a<b):
        plt.savefig('saddle_point_1b.png')
    if (a>b):
        plt.savefig('nodo_inestable_1b.png')

    
fig, fig6=plt.subplots(1,1, figsize=(8,4))
campoB(1, 2, figx=fig6)

```

# APARTADO C

```python
#hacemos un plot
def plot1(a, b, c, d, figx=None):
    L = np.linspace(-2, 2)
    R = np.linspace(-2, 2)
    
    figx.plot(R, - a * R / b, 'k', label=(r'$\dot R =0$'))
    figx.plot(R, - (d / c) * R * (1 - R ** 2), 'b', label=(r'$\dot L =0$'))
    figx.plot([-2, 0, 2], [2, 0, -2], 'ok', label=('Puntos Fijos'))
    
fig, fig4=plt.subplots(1, 1, figsize=(8,4))

plot1(1, 1, -3, 1, figx=fig4)
plt.legend()
plt.savefig('nulclina_1c.png')
```

```python
#campo vectorial c
def campoC(figx=None):
    R1 = np.linspace(-6, 6, 100)
    
    L1 = - R1
    L2 =  (1 / 3) * R1 * (1 - (R1 ** 2))
    
    figx.plot(R1, L1, 'k', label=(r'$\dot R =0$'))
    figx.plot(R1, L2, 'b', label=(r'$\dot L =0$'))
    figx.plot([-2, 0, 2], [2, 0, -2], 'ok', label=('Puntos Fijos'))

    
    figx.set_xlabel('R')
    figx.set_ylabel('L')
    figx.set_xlim(-4, 4)
    figx.set_ylim(-4, 4)
    #figx.set_title(r'$x_+^*$')
    
    
    R, L = np.meshgrid(np.linspace(-4, 4, 50), np.linspace(-4, 4, 50))
    
    Rpunto = R + L
    Lpunto = - (3 * L) + (R * (1 - (R ** 2)))

    
    
    #figx.quiver(R, L, Rpunto, Lpunto, angles='xy', pivot='mid', scale=50, scale_units='x', headwidth=4, minshaft=0.9)
    figx.streamplot(R, L, Rpunto, Lpunto, density=1.5)

    
    
fig, fig7=plt.subplots(1,1, figsize=(8,4))
campoC(figx=fig7)
plt.savefig('campo_1c.png')
```
